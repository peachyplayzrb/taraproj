```python
def combine_fields(row):
    # Title repeated twice to give it natural weight boost
    title = str(row['title']).strip()
    abstract = str(row['abstract']).strip()
    return f"{title} {title} {abstract}"

df_cs['combined_text'] = df_cs.apply(combine_fields, axis=1)
```

```python
def get_pos_weights(text):
    tokens = word_tokenize(text.lower())
    tagged = nltk.pos_tag(tokens)

    weights = {}
    for word, tag in tagged:
        if '-' in word:                   # Hyphenated compound terms
            weights[word] = 1.8
        elif tag.startswith('NN'):        # Nouns
            weights[word] = 2.0
        elif tag.startswith('VB'):        # Verbs
            weights[word] = 1.2
        elif tag.startswith('JJ'):        # Adjectives
            weights[word] = 1.0
        else:
            weights[word] = 0.5
    return weights
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import scipy.sparse as sp

# Build base TF-IDF matrix from your cleaned, stemmed text
vectorizer = TfidfVectorizer(
    max_features=50000,    # cap dimensions to keep it manageable
    min_df=2,              # ignore terms appearing in fewer than 2 docs
    max_df=0.95,           # ignore terms appearing in more than 95% of docs
    ngram_range=(1, 2)     # include bigrams e.g. "neural network" as one feature
)

tfidf_matrix = vectorizer.fit_transform(df_cs['processed_text'])
feature_names = vectorizer.get_feature_names_out()

print(f"Matrix shape: {tfidf_matrix.shape}")
print(f"Documents: {tfidf_matrix.shape[0]}, Features: {tfidf_matrix.shape[1]}")
print(f"Sample features: {feature_names[:20]}")
```

```python
def build_pos_weight_vector(feature_names, df_pos_weights):
    """
    For each term in the vocabulary, find its average POS weight
    across all documents that contain it.
    """
    # Aggregate weights across all documents
    term_weights = {}
    term_counts = {}

    for doc_weights in df_cs['pos_weights']:
        for term, weight in doc_weights.items():
            # Stem the term to match the vocabulary
            stemmed = stemmer.stem(term)
            if stemmed in term_weights:
                term_weights[stemmed] += weight
                term_counts[stemmed] += 1
            else:
                term_weights[stemmed] = weight
                term_counts[stemmed] = 1

    # Build weight vector aligned to feature_names order
    pos_weight_vector = np.ones(len(feature_names))  # default = 1.0
    for i, term in enumerate(feature_names):
        if term in term_weights:
            pos_weight_vector[i] = term_weights[term] / term_counts[term]

    return pos_weight_vector

pos_weight_vector = build_pos_weight_vector(feature_names, df_cs['pos_weights'])

print(f"POS weight vector shape: {pos_weight_vector.shape}")
print(f"Min weight: {pos_weight_vector.min():.2f}")
print(f"Max weight: {pos_weight_vector.max():.2f}")
print(f"Mean weight: {pos_weight_vector.mean():.2f}")
```

```python
# Multiply each column (term) by its POS weight
# tfidf_matrix is sparse — use efficient sparse multiplication
pos_weight_diag = sp.diags(pos_weight_vector)
tfidf_pos_matrix = tfidf_matrix.dot(pos_weight_diag)

print(f"POS-weighted matrix shape: {tfidf_pos_matrix.shape}")

# Verify a sample — compare a term's weight before and after
sample_term = 'neural'
if sample_term in feature_names:
    idx = list(feature_names).index(sample_term)
    print(f"\nTerm '{sample_term}':")
    print(f"  POS multiplier: {pos_weight_vector[idx]:.3f}")
    print(f"  TF-IDF before: {tfidf_matrix[0, idx]:.4f}")
    print(f"  TF-IDF after:  {tfidf_pos_matrix[0, idx]:.4f}")
```

```python
def search(query, top_k=10):
    """
    Process query through the same pipeline as documents,
    apply POS weighting, compute cosine similarity, return ranked results.
    """
    # Step 1 — preprocess query (same pipeline as documents)
    query_clean = query.lower()
    query_clean = remove_punctuation(query_clean)
    query_clean = re.sub(r'\d+', '', query_clean)
    query_tokens = word_tokenize(query_clean)
    query_tokens = [t for t in query_tokens if t not in stop_words]
    query_tokens = [stemmer.stem(t) for t in query_tokens]
    query_processed = ' '.join(query_tokens)

    # Step 2 — vectorise using the same fitted vectorizer
    query_vec = vectorizer.transform([query_processed])

    # Step 3 — apply POS weights to query vector
    query_vec_pos = query_vec.dot(pos_weight_diag)

    # Step 4 — cosine similarity against all documents
    from sklearn.metrics.pairwise import cosine_similarity
    scores = cosine_similarity(query_vec_pos, tfidf_pos_matrix).flatten()

    # Step 5 — rank and return top k
    top_indices = scores.argsort()[::-1][:top_k]

    results = []
    for rank, idx in enumerate(top_indices, 1):
        results.append({
            'rank': rank,
            'id': doc_ids[idx],
            'title': doc_titles[idx],
            'abstract': doc_abstracts[idx][:300] + '...',  # snippet
            'score': round(scores[idx], 4)
        })

    return results
# Test it
results = search("deep learning image classification")
for r in results:
    print(f"{r['rank']}. [{r['score']}] {r['title']}")
    print(f"    {r['abstract'][:150]}...")
    print()
```

```python
def search_baseline(query, top_k=10):
    """Same search but using base TF-IDF without POS weighting"""
    query_clean = query.lower()
    query_clean = remove_punctuation(query_clean)
    query_clean = re.sub(r'\d+', '', query_clean)
    query_tokens = word_tokenize(query_clean)
    query_tokens = [t for t in query_tokens if t not in stop_words]
    query_tokens = [stemmer.stem(t) for t in query_tokens]
    query_processed = ' '.join(query_tokens)

    query_vec = vectorizer.transform([query_processed])
    from sklearn.metrics.pairwise import cosine_similarity
    scores = cosine_similarity(query_vec, tfidf_matrix).flatten()
    top_indices = scores.argsort()[::-1][:top_k]

    return [{'rank': i+1, 'id': doc_ids[idx],
              'title': doc_titles[idx], 'score': round(scores[idx], 4)}
            for i, idx in enumerate(top_indices)]

# Side by side comparison
query = "neural network optimisation"
print(f"Query: '{query}'\n")
print("=== BASELINE TF-IDF ===")
for r in search_baseline(query, top_k=5):
    print(f"  {r['rank']}. [{r['score']}] {r['title']}")

print("\n=== POS-WEIGHTED TF-IDF ===")
for r in search(query, top_k=5):
    print(f"  {r['rank']}. [{r['score']}] {r['title']}")
```

```python
# Click-through re-ranking store
# Tracks how many times a document was clicked for a given query

from collections import defaultdict

# Structure: {normalised_query_string: {doc_id: click_count}}
click_store = defaultdict(lambda: defaultdict(int))

def normalise_query(query):
    """Normalise query the same way as search pipeline for consistent keys"""
    q = query.lower()
    q = remove_punctuation(q)
    q = re.sub(r'\d+', '', q)
    tokens = word_tokenize(q)
    tokens = [t for t in tokens if t not in stop_words]
    tokens = [stemmer.stem(t) for t in tokens]
    return ' '.join(sorted(tokens))  # sorted so "learning deep" == "deep learning"

def record_click(query, doc_id):
    """Call this when a user clicks a result"""
    key = normalise_query(query)
    click_store[key][doc_id] += 1
    print(f"Recorded click: query='{key}' doc='{doc_id}'")

def get_click_score(query, doc_id):
    """Returns normalised click score for a doc given a query"""
    key = normalise_query(query)
    clicks = click_store[key]
    if not clicks or doc_id not in clicks:
        return 0.0
    max_clicks = max(clicks.values())
    return clicks[doc_id] / max_clicks  # normalise to 0-1

print("Click store initialised")
```

```python
def search_with_clicks(query, top_k=10, alpha=0.7):
    """
    Blended ranking: alpha * cosine_score + (1-alpha) * click_score
    alpha=0.7 keeps retrieval quality dominant but surfaces popular results
    """
    # Step 1 — get base cosine scores
    query_clean = query.lower()
    query_clean = remove_punctuation(query_clean)
    query_clean = re.sub(r'\d+', '', query_clean)
    query_tokens = word_tokenize(query_clean)
    query_tokens = [t for t in query_tokens if t not in stop_words]
    query_tokens = [stemmer.stem(t) for t in query_tokens]
    query_processed = ' '.join(query_tokens)

    query_vec = vectorizer.transform([query_processed])
    query_vec_pos = query_vec.dot(pos_weight_diag)

    from sklearn.metrics.pairwise import cosine_similarity
    cosine_scores = cosine_similarity(query_vec_pos, tfidf_pos_matrix).flatten()

    # Step 2 — blend with click scores
    blended_scores = []
    for idx in range(len(doc_ids)):
        doc_id = doc_ids[idx]
        cosine = cosine_scores[idx]
        click = get_click_score(query, doc_id)
        blended = alpha * cosine + (1 - alpha) * click
        blended_scores.append(blended)

    blended_scores = np.array(blended_scores)
    top_indices = blended_scores.argsort()[::-1][:top_k]

    results = []
    for rank, idx in enumerate(top_indices, 1):
        doc_id = doc_ids[idx]
        click_s = get_click_score(query, doc_id)
        results.append({
            'rank': rank,
            'id': doc_id,
            'title': doc_titles[idx],
            'abstract': doc_abstracts[idx][:300] + '...',
            'cosine_score': round(cosine_scores[idx], 4),
            'click_score': round(click_s, 4),
            'blended_score': round(blended_scores[idx], 4)
        })

    return results
```

```python
# Simulate a user scenario — demonstrate click re-ranking working

query = "graph neural network node classification"

print("=== BEFORE ANY CLICKS ===")
results_before = search_with_clicks(query, top_k=5)
for r in results_before:
    print(f"  {r['rank']}. [{r['blended_score']}] {r['title'][:60]}")

# Simulate: user searches, sees results, clicks the 3rd result twice
clicked_doc_id = results_before[2]['id']
clicked_doc_title = results_before[2]['title']
record_click(query, clicked_doc_id)
record_click(query, clicked_doc_id)
print(f"\nSimulated 2 clicks on: '{clicked_doc_title[:60]}'")

print("\n=== AFTER CLICKS — same query ===")
results_after = search_with_clicks(query, top_k=5)
for r in results_after:
    marker = " <-- boosted" if r['id'] == clicked_doc_id else ""
    print(f"  {r['rank']}. [{r['blended_score']}] (cos:{r['cosine_score']} | click:{r['click_score']}) {r['title'][:50]}{marker}")
```

```python
# Alpha sensitivity analysis
query = "graph neural network node classification"

# Use the correct variable name from cell 37
record_click(query, clicked_doc_id)
record_click(query, clicked_doc_id)

print(f"{'Alpha':<8} {'Rank-1 Title':<55} {'Boosted doc rank'}")
print("-" * 80)
for alpha in [0.5, 0.6, 0.7, 0.8, 0.9]:
    top = search_with_clicks(query, top_k=10, alpha=alpha)
    rank1 = top[0]['title'][:50].replace('\n', ' ')
    boosted_rank = next((r['rank'] for r in top if r['id'] == clicked_doc_id), 'N/A')
    print(f"{alpha:<8} {rank1:<55} {boosted_rank}")
```

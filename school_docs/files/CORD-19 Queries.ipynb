# Query demo

See also __[Query API](https://www.elastic.co/guide/en/elasticsearch/reference/current/query-dsl-simple-query-string-query.html)__

```python
from elasticsearch import Elasticsearch

# create an Elasticsearch client instance
es = Elasticsearch("http://localhost:9200/")

# define the index to query
index_name = "cord19-2"

# define the search query
#search_query = {
#    "match": {
#      "title": "covid vaccine"
#    }
#}
search_query = {
        "multi_match": {
            "query": "covid vaccine",
            "fields": ["title^3", "abstract"]
        }
  
}

# perform the search
result = es.search(index=index_name, query=search_query)

# print the results
maxscore = result['hits']['max_score']
print("maxscore:", maxscore) # use to normalise scores

print("Received %d documents." % result['hits']['total']['value'])

for hit in result['hits']['hits']:
    print(hit['_score']/maxscore, hit['_source']['authors'], hit['_source']['title'])

```
